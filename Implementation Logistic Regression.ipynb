```python
def cost(y_pred,y):
    y_pred=np.array(y_pred)
    y=np.array(y)
    COST=-y*np.log(y_pred)-(1-y)*np.log(1-y_pred)
    return np.sum(COST)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def GradientStep(X,y,parameters):
    z=np.matmul(X,parameters)
    #print(z)
    y_pred=sigmoid(z)
    diff=y_pred-y
    changes=np.matmul(X.T,diff)
    return changes
```

```python
def GradientDescent(X,y,epochs,learning_rate):
    np.random.seed(0)
    parameters=(np.random.randn(X.shape[1])/10).reshape(-1,1)
    for i in range(epochs):
        changes=GradientStep(X,y,parameters)
        parameters=parameters-learning_rate*changes
    return parameters
```

```python
def LogisticRegressionFit(X,y,epochs=100,learning_rate=0.001):
    Ones=np.ones(X.shape[0]).reshape(-1,1)
    X=np.concatenate((Ones,X),axis=1)
    Y=pd.get_dummies(y[:,0])
    Y=Y.values
    parameters=[]
    for i in range(Y.shape[1]):
        temp_parameters=GradientDescent(X,Y[:,i],epochs,learning_rate)
        parameters.append(temp_parameters[:,0])
    parameters=np.array(parameters).T
    return parameters
```

```python
def LogisticPredict(X_train,y_train,X,lr=0.001):
    parameters=LogisticRegressionFit(X_train,y_train,learning_rate=lr)
    Ones=np.ones(X.shape[0]).reshape(-1,1)
    X=np.concatenate((Ones,X),axis=1)
    z=np.matmul(X,parameters)
    y_pred=sigmoid(z)
    #print(y_pred)
    #print(np.argmax(y_pred,axis=1))
    return y_pred
```

```python
LRlist=[ 0.00001*np.power(2,i) for i in range(16)]
def Best_lr(X_train,y_train,lr_matrix=LRlist):
    MinCost=sys.maxsize
    blr=0
    for i in lr_matrix:
        temp_pred=LogisticPredict(X_train,y_train,X_train)
        temp_cost=cost(temp_pred,y)
        if(temp_cost<MinCost):
            MinCost=temp_cost
            blr=i
    #print(MinCost)
    return blr
```
